<h1>what is Matplotlib?</h1>

Matplotlib is an amazing visualization library in Python for 2D plots of arrays. Matplotlib is a multi-platform data visualization library built on NumPy arrays and designed to work with the broader SciPy stack. It was introduced by John Hunter in the year 2002.

One of the greatest benefits of visualization is that it allows us visual access to huge amounts of data in easily digestible visuals. Matplotlib consists of several plots like line, bar, scatter, histogram etc.

<h1>Pyplot</h1>

Pyplot is a collection of command style functions that makes some change to a figure: e.g., creates a figure, creates a plotting area in a figure, plots some lines in a plotting area, decorates the plot with labels, etc. In Pyplot various states are preserved across function calls, so that it keeps track of things like the current figure and plotting area, and the plotting functions are directed to the current axes (please note that “axes” here and in most places in the documentation refers to the axes part of figure and not the strict mathematical term for more than one axis).

<h2>a. plot()</h2>

You can use the plot() method to create a plot of points on the graph.


```python
import matplotlib.pyplot as plt
plt.plot([2,3,4,5])
plt.xlabel('Label on x-axis')   #for label on x-axis
plt.ylabel('Label on y-axis')   #for label on y-axis
plt.show()
```

<h2>b. Formatting your Python Plot</h2>
<ul>
<li>A third argument will let you choose the color and the line type of the plot in Python Programming Language.</li>
<li>The default format string gives us a solid blue line, as we’ve seen in the examples so far. This is ‘b-‘. You see, these strings are from MATLAB.</li>

```python
import matplotlib.pyplot as plt
import numpy as np
t=np.arange(0,5,0.2)
plt.plot(t,t,'r--',t,t**3,'b^',t,t**2,'gs')
plt.show()
```

<h1>Matplotlib Keyword Strings</h1>

Using a data keyword argument, we can produce Python plots with strings that relate to some variables.
We can access these variables with strings.

```python
import matplotlib.pyplot as plt
import numpy as np
data={'a':np.arange(50),'c':np.random.randint(0,50,50),'d':np.random.randn(50)}
data['b']=data['a']+10*np.random.randn(50)
data['d']=np.abs(data['d'])*100
plt.scatter('a','b',c='c',s='d',data=data)
plt.show()
```

<h1>Categorical Variables to Python Plotting</h1>

We can pass categorical variables to Python plotting functions.


```python
import matplotlib.pyplot as plt
names=["Dingos","Wild Cats","Tigers"]
values=[1,11,111]
plt.figure(1,figsize=(9,3))
plt.subplot(131)
plt.bar(names,values)
plt.subplot(132)
plt.scatter(names,values)
plt.subplot(133)
plt.plot(names,values)
plt.suptitle('Varsity')
plt.show()
```

<h1>Matplotlib Keyword Strings</h1>

Using some keyword arguments, we can alter how a graph looks. Let’s look at some.



<h2>a. Linewidth</h2>

By using linewidth we can alter the line width


```python
import matplotlib.pyplot as plt
plt.plot([1,2,3],[2,4,9],linewidth=4.0)
plt.show()
```

<h2>b. Alpha</h2>

We can use the alpha channel to create prettier plots by softening colors.


```python
import matplotlib.pyplot as plt
plt.plot([1,2,3],[2,4,9],alpha=5.5)
plt.show()
```

<h2>c. Antialiased</h2>
<ul>
<li>If you’ll look closely, the lines look quite smooth.</li>
<li>But we can turn antialiasing off- this will show us the aliasing in the lines.</li></ul>


```python
import matplotlib.pyplot as plt
plt.plot([1,2,3],[2,4,9],antialiased=True)
plt.show()
```

<h2>d. Color or c</h2>

You can set the color of the plot with this parameter.


```python
import matplotlib.pyplot as plt
plt.plot([1,2,3],[2,4,9],color='Chartreuse')
plt.show()
```

<h2>e. Dashes</h2>

Helps to add dashed lines into plots.


```python
import matplotlib.pyplot as plt
plt.plot([1,2,3],[2,4,9],dashes=[1,2,4,4])
plt.show()
```

<h2>f. Linestyle or ls</h2>

You can choose the style of line you want for your plot.


```python
import matplotlib.pyplot as plt
plt.plot([1,2,3],[2,4,9],linestyle='steps')
plt.show()
```

<h2>g. Marker</h2>

A marker will let you select what symbol you want to display at breakpoints and bends.


```python
import matplotlib.pyplot as plt
plt.plot([1,2,3],[2,4,9],marker='+')
plt.show()
```

<h2>h. Markeredgecolor</h2>

You can also decide on a color for your markers’ edges.


```python
import matplotlib.pyplot as plt
plt.plot([1,2,3],[2,4,9],marker='+',markeredgecolor='brown')
plt.show()
```

<h2>i. Markeredgewidth</h2>

This lets us decide how thick we want our markers to be.


```python
import matplotlib.pyplot as plt
plt.plot([1,2,3],[2,4,9],marker='+',markeredgewidth=0.4)
plt.show()
```

<h2>j. Markerfacecolor and Markersize</h2>

This will let you choose which color to fill in your marker and what size to keep it.


```python
import matplotlib.pyplot as plt
plt.plot([1,2,3],[2,4,9],marker='.',markerfacecolor='orange',markersize=13.0)
plt.show()
```

<h2>k. Markevery</h2>

This parameter lets us decide at what parameters we want to put markers.


```python
import matplotlib.pyplot as plt
plt.plot([1,2,3],[2,4,9],marker='.',markerfacecolor='orange',markersize=13.0,markevery=2)
plt.show()
```

<h2>l. Zorder</h2>

This lets us decide which plots will show up in the front and which to send to back.


```python
import matplotlib.pyplot as plt
plt.plot([1,2,3],[2,4,9],zorder=1,linewidth=4)
plt.plot([1,2,6,9],[2,4,9,10],zorder=2,linewidth=4)
plt.show()
```

<h2> Grid in Python Plot</h2>

You can use the grid() method to toggle a grid in your plot.


```python
import matplotlib.pyplot as plt
plt.grid(True)
plt.plot([1,2,6,9],[2,4,9,10],zorder=2,linewidth=4)
plt.show()
```
